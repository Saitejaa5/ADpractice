```python
!pip install seaborn

import pandas as pd
import numpy as np
df1 = pd.read_csv(r"C:\Users\Gaurav\Downloads\universities.csv")
print(df1.head())
data = df1['SAT']
print(data.head())
```

```python
mean = np.mean(data)
median = np.median(data)
var = np.var(data)
std = np.std(data)
rnge = np.max(data) - np.min(data)
```

```python
print(f"Mean of SAT Scores : {mean}")
print(f"Median of SAT Scores : {median}")
print(f"Variance of SAT Scores : {var}")
print(f"Range of SAT Scores : {rnge}")
print(f"Standard Deviation of SAT Scores : {std}")
```
