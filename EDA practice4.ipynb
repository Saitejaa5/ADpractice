```python

print(data1['Month'].value_counts())
mode_month = data1["Month"].mode()[0]
print(mode_month)
```

```python
print(data1['Solar.R'].value_counts())
mode_Solar = data1["Solar.R"].mode()[0]
print(mode_Solar)
```

***Dectection of outliers in the columns***


```python


# Select numeric columns only
numeric_cols = data1.select_dtypes(include=[np.number])

# Plot histograms and box plots for each numeric column
num_cols = len(numeric_cols.columns)
fig, axes = plt.subplots(num_cols, 2, figsize=(12, 5 * num_cols))

for i, col in enumerate(numeric_cols.columns):
    # Histogram
    sns.histplot(data1[col], kde=True, ax=axes[i, 0])
    axes[i, 0].set_title(f'Histogram of {col}')
    axes[i, 0].set_xlabel(col)
    axes[i, 0].set_ylabel('Frequency')
    
    # Boxplot
    sns.boxplot(x=data1[col], ax=axes[i, 1])
    axes[i, 1].set_title(f'Boxplot of {col}')
    axes[i, 1].set_xlabel(col)

plt.tight_layout()
plt.show()
```

***Method1 : using histogram and boxplots***

```python
fig,axes=plt.subplots(2,1,figsize=(8,6),gridspec_kw={'height_ratios':[1,3]})
sns.boxplot(data=data1["Ozone"],ax=axes[0],color='skyblue',width=0.5,orient='h')
axes[0].set_title("Boxplot")
axes[0].set_xlabel("Ozone Levels")

sns.histplot(data1["Ozone"],kde=True,ax=axes[1],color='Purple',bins=30)
axes[1].set_title("Histogram with KDE")
axes[1].set_xlabel("Ozone Levels")
axes[1].set_ylabel("Frequency")
plt.tight_layout()
plt.show()
```
